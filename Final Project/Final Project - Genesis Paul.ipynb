## Inflation

This project is going to be especially difficult, because all of my key variables are impacted by inflation, thus will all likely be highly correlated with one another. However, for this portion in particular I want to observe what federal minimum wage would look like had it followed inflation.

I will be using the value of the dollar and making an assumption that the value of the dollar in 1968 was \\$1.00. I understand this assumption is in fact wrong, but it will allow us to be able to best visualize how federal minimum wage would have changed in comparison to how it has.

I will be using a line graph to see how the two compare.

```python
inflation = pd.read_csv("Inflation.csv")
inflation = inflation.rename(columns = {"year": "Year",
                                       "amount": "Dollar Value",
                                       "inflation rate": "Inflation Rate"}).set_index(["Year"]).loc[1968:2020]
inflation["Dollar Value"] = (inflation["Dollar Value"] / 12.1).round(2) #shifts 1995 as the base year 
inflation
```

```python
inflated_data = data.merge(inflation, left_index=True, right_index=True)
inflated_data = inflated_data.reset_index()
inflated_data = inflated_data.drop(columns = ['State'])
inflated_data = inflated_data.set_index(['Year']).groupby(level = 0).mean()
inflated_data
```

```python
inflated_data["Inflated Wage"] = " "
inflated_data["Inflated Wage"].loc[1968] = inflated_data["Federal Minimum Wage"].loc[1968]
inIndex = inflated_data.index
inIndex = inIndex.drop(1968)
for i in inIndex:
    inflated_data["Inflated Wage"].loc[i] = (inflated_data["Inflated Wage"].loc[i-1] * (1+inflated_data["Inflation Rate"].loc[i-1])).round(2)
inflated_data
```

*Line Graph*

Comparison of actual federal minimum wage and where it would be if 1968's minimum wage is assumed to be the base value of a dollar and inflated respectively

We will see that the inflated wage has surpassed the actual federal minimum wage since the 1980's and that the gap is continuing to grow. Thus, those in favor of minimum wage are correct in stating it hasn't kept up with inflation, and had I not assumed 1968 were the base we would expect the actual gap to be even larger. Thus, calling for an increase.

```python
fig, ax = plt.subplots(figsize = (16, 12))
inflated_data.drop(columns=["Effective Minimum Wage", "Dollar Value", "Inflation Rate"]).plot(ax = ax,
                                                                                             title = "Federal Minimum Wage Accounting for Inflation")

y_vals = ax.get_yticks()
ax.set_yticklabels(["$" + str(int(y)) + ".00" if y >= 1 else round(y,1) for y in y_vals]) 
```

**Congregated Unemployment Data**

Using the rates and unemployment data I collected I will be calculating the number of people actively working, the total number of people in the workforce, the population, and the number of people that are not accounted in the workforce. (These values do not include children, but does include adults that are not in the work force for a variety of reasons)

```python
unemploy = pd.concat([unemploy_rate, participation, unemploy_level], axis = 1, join="inner")
unemploy
```

```python
unemploy["Working"] = (unemploy["Unemployed"] / (unemploy["Unemployment Rate"]/100)) - unemploy["Unemployed"]
unemploy["Workforce"] = unemploy["Unemployed"] / (unemploy["Unemployment Rate"]/100)
unemploy["Not in Workforce"] = (unemploy["Workforce"] / (unemploy["Participation Rate"]/100)) - unemploy["Workforce"]
unemploy["Population"] = unemploy["Workforce"] / (unemploy["Participation Rate"]/100)
unemploy_data = unemploy.drop(columns = ["Unemployment Rate", "Participation Rate"])
unemploy_data
```

## Poverty
I collected data regarding the poverty rate in the us from 1989 - 2020, as this was the earliest data I had. I will be using the population I established in the unemployment section * the poverty rate to establish the count of people that are living in poverty.

In my analysis of poverty, I will be including unemployment data as I am making an assumption that they are associated with one another. I am doing this because if the unemployment distribution is associated with poverty I can draw further conclusions about federal minimum wage.

```python
poverty = pd.read_csv("Poverty.csv")
poverty = poverty.rename(columns = {"DATE": "Year",
                                   "PPAAUS00000A156NCEN": "Poverty Rate"}).set_index(["Year"]).loc[1968:2020]
poverty
```

```python
poverty_data = pd.concat([unemploy_data, poverty], axis = 1, join="inner")
poverty_data["Poverty"] = poverty_data["Population"] * (poverty_data["Poverty Rate"] / 100)
poverty_data
```

*Correlation Map*

We will use this the below correlation map to draw further conclusions about both poverty and unemployment. 

We will see that there is almost no correlation between the unemployed and the working class. Thus, these would make acceptable independent variables in an OLS regression where poverty is the dependent variable.

```python
povUn_corr = corr_matrix_heatmap(poverty_data.drop(columns = ["Workforce", "Population", "Poverty Rate"]).corr(),
                                  save_fig = False,
                                  pp = None,
                                  title = "Poverty & Unemployment Correlation")
povUn_corr
```

```python
import statsmodels.api as sm
```

*Regression*

Ho: There is no statistical significance between the the poverty and the number of unemployed or those that are actively working.

```python
y_var = ["Poverty"]
x_vars = ["Unemployed",
         "Working"]
reg_vars = y_var + x_vars
reg_data = poverty_data[reg_vars].dropna()
reg_data
```

```python
y = reg_data[y_var]
X = reg_data[x_vars]
X["Constant"]=1
results = sm.OLS(y,X).fit()
results.summary()
```

```python
predictor = results.predict()
reg_data[y_var[0] + " Predictor"] = predictor
reg_data.plot(figsize = (18, 8),
             title = "Poverty Predictor")
```

With an acceptable but not exceptional $R^2$ value we see that the working class (unemployed and working) reasonably account for the poverty level. And based on our P value, we can state that at a significant level of 90% we can reject the null hypothesis and there is statistical significane between the working class and poverty, thus we can adequately predict poverty.

### Poverty Conclusion

Unemployment directly impacts poverty and if we can reduce the unemployment level, we may be able to reduce the poverty level as well. I also am assuming that to reduce unemployment, we must increase minimum wage.
