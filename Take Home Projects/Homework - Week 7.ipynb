```python
def gather_data(data_codes, 
                start, 
                end = datetime.datetime.today(), 
                freq = "M"):
    i = 0

    for key, val in data_codes.items():
        if i == 0:

            df = web.DataReader(
                val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 

            i = None
        else:

            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()

    return df
```

##### Scatter Plot

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def color_dim_scatter(data, pp):
    for x in data:
        for y in data:
            if x != y:
                for c in data:
                    if x != c and y != c:
                        fig, ax = plt.subplots(figsize = (12,12))
                        data.plot.scatter(x=x,
                                         y=y,
                                         c=c,
                                         alpha=.5,
                                         colormap="inferno",
                                         ax=ax)
                        plt.xticks(rotation = 90)
                        plt.setp(ax.get_xticklines(), visible =False)
                        plt.setp(ax.get_yticklines(), visible=False)
                        plt.show()
                        pp.savefig(fig, bbox_inches="tight")
                        plt.close()
                        
    
pp = PdfPages("Unemployment Plots.pdf")
plt.rcParams.update({"font.size":22})
color_dim_scatter(data,pp)
pp.close()
```

##### Line Plot

```python
rate = data[["Unemployment Rate", "Natural Unemployment Rate"]]

fig, ax = plt.subplots(figsize = (16, 8))
rate.plot(ax = ax,
          legend = True)

plt.title("Unemployment")
```

##### Correlation Matrix

```python
import seaborn as sn
corr = data.corr()
sn.heatmap(corr, annot = True)
plt.show()
```

### Compare Data

> Line Plot: There appears to be no direct correlation between unemployment rate and natural unemployment rate. The line for natural employment goes directly through the center of the unemployment rate.

> Correlation: The unemployment rate is positively correlated to the unemployment level. The natural unemployment rate is negatively corrrelated to unemployment level.

### Create a correlation matrix plot for quarterly observations of Nominal GDP, Real GDP, the Price Level (CPI or GDP Deflator), the quantity of currency in circulation and the federal funds rate for as large of a date range as possible. Then, calculate the year-over-year percent change for each each variable and save the results as variables in the same dataframe. Create a correlation matrix for these new variables.

```python
real_gdp = pd.read_csv("Real GDP.csv")
nominal_gdp = pd.read_csv("Nominal GDP.csv")
cpi = pd.read_csv("CPI.csv")
```

```python
gdp = pd.merge(real_gdp, nominal_gdp, on = "DATE")
data = pd.merge(gdp, cpi, on = "DATE")

data
```

```python
data.keys()
```

```python
data = data.rename(columns = {"GDPC1": "Real GDP", "NGDPPOT": "Nominal GDP", "CPALTT01USM657N": "CPI"})
data
```

```python
real_arr = []
for i in range(243):
    value = ((data["Real GDP"].iloc[i] - data["Real GDP"].iloc[i+3]) / (data["Real GDP"].iloc[i])) * 100
    real_arr.append(value)

real_arr.append(0)
real_arr.append(0)
real_arr.append(0)
```

```python
nom_arr = []
for i in range(243):
    value = ((data["Nominal GDP"].iloc[i] - data["Nominal GDP"].iloc[i+3]) / (data["Nominal GDP"].iloc[i])) * 100
    nom_arr.append(value)

nom_arr.append(0)
nom_arr.append(0)
nom_arr.append(0)

nom_arr
```

```python
cpi_arr = []
for i in range(243):
    value = ((data["CPI"].iloc[i] - data["CPI"].iloc[i+3]) / (data["CPI"].iloc[i])) * 100
    cpi_arr.append(value)
    
cpi_arr.append(0)
cpi_arr.append(0)
cpi_arr.append(0)
```

```python
data["Real % Change"] = real_arr
data["Nominal % Change"] = nom_arr
data["CPI % Change"] = cpi_arr
```
